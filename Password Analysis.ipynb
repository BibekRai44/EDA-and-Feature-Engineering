## Basic Data Visualization

```python
x1=data['password'].value_counts().sum()
```

```python
x=([x1,x2])
mylabels=['password count','strength count']
myexplode=[0.2,0.4]
plt.pie(x,labels=mylabels,autopct='%1.2f%%',shadow=True,explode=myexplode)
```
